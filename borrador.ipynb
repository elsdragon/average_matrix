## (1)

Escribe la función `sum_all` que recibe una lista de números y devuelve la suma total.

En este ejercicio, veremos el proceso general que hay que seguir y los demás los harás tú sólo.

```python
def sum_all(numbers):
  """
  Recibe una lista de números y devuelve su total.
  Si recibe la lista vacia, devuelve 0
  """
  # Creo una variable acumulador y le doy un valor inical adecuado
  total = 0
  # Recorro la lista sumando al valor actual al acumulador
  for number in numbers:
    total = total + number

  # Cuando llegue al final tendré en el acumulador el total de todos los elementos
  return total
```

```python
lista = (1, 2, 3, 4)
sum_all(lista)
```

```python
def add_none_matrix(matrix):
    
    new_matrix = matrix
    
    for i in range(len(new_matrix)):
        new_matrix[i].insert(0, None)
        new_matrix[i].append(None)
        

    new_list = []
    i = 0
    while i < len(matrix[0]):
        new_list.append(None)
        i += 1
    
    new_matrix.insert(0,new_list)
    new_matrix.append(new_list)
    
    return new_matrix
```

```python
def add_list_neighbour(matrix):
    ''' create a new list with neighbour elements on matrix'''
    neighbour_matrix = []

    for i in range(1, len(matrix)-1):
        new_matrix= []
        for j in range(1, len(matrix[i])-1):
            new_list = []
            new_list.append(matrix[i][j])
            new_list.append(matrix[i-1][j])
            new_list.append(matrix[i+1][j])
            new_list.append(matrix[i][j-1])
            new_list.append(matrix[i][j+1])
            new_matrix.append(new_list)

        neighbour_matrix.append(new_matrix)    
    
    return neighbour_matrix
```

```python
def medium_of_list(list):
    ''' 
    Calculate arithmetic average from a list'''
    total = 0
    divisor = len(list)
    for element in list:
        if element != None:
            total += element
        if element == None:
            divisor -= 1
    
    return total/divisor
```
